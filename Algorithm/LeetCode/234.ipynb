### [LeetCode]234. Palindrome Linked List
***
- **Problem**  
***
> **Given a singly linked list, determine if it is a palindrome.**  
***
> ***Example 1***  
> **Input: 1->2  
> Output: false**
***
> ***Example 2:***  
> **Input: 1->2->2->1  
> Output: true**
 
***
> ***Follow up:***  
> **Could you do it in O(n) time and O(1) space?**
***
- **Solve**
__Runtime: 60 ms, faster than 96.39% of Python3 online submissions for Palindrome Linked List.__

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head):
        q = collections.deque()
        
        if not head :
            return True
        
        node = head
        while node is not None:
            q.append(node.val)
            node = node.next
        
        while len(q) > 1:
            if q.popleft() != q.pop():
                return False
            
        return True
```
